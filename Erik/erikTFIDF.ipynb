We see 4545 unique terms, and know that there are 394 docs, so expect to generate a 4545x394 pivot table.

Here, I will use a basic formula to calculate, as this is easily changed if necessary. 
First, note the formula
$$
tfidf(t, d, D) = tf(t,d)\cdot idf(t,D)
$$
where $t$ is the term, $d$ is the document, and $D$ is the corpus. 
Then, apply the raw count TF weight, and the inverse doc freq weight of 
$$
log\frac{N}{n_t}
$$
where $N = |D|$ and n_t is the number of docs where the term appears (formally $|{d \in D: t \in d}|$)


this means the overall expression for tfidf is
$$
f_{t,d}\cdot ln(\frac{N}{n_t})
$$

```python
pivotcount = alltex.pivot_table(aggfunc=pd.Series.nunique, columns='doc', index='Term', values= 'form', dropna=False)
pivotcount
```

Terms that are in a list of unique values for Term column, but did not show up in pivot table.

```python
notin = pd.Series.unique(alltex.Term)[~np.in1d(pd.Series.unique(alltex.Term), pivotcount.index)]
```

```python
alltex.loc[np.in1d(alltex.Term, notin), :]
```

We now see that these terms weren't included in the pivot table because they have no doc entry (thus id_text was probably also NaN).  Let's try it with titles instead.

```python
pivottitlescount = alltex.pivot_table(aggfunc=pd.Series.nunique, columns='text_name', index='Term', values= 'form', dropna=False)
pivottitlescount
```

This looks more complete, with all 4545 terms accounted for. Now, we have count per document, so each of these cells is a particular $f_{t,d}$

To calculate log(N/n_t), we can simply note that N is 394, and n_t is the number of not NaN entries per row. This is easily done.

```python
def idf(row):  
    return row * np.sum(~np.isnan(pivottitlescount.iloc[1]))/len(row)
```

```python
tfidftable = pivottitlescount.apply(func=idf)
```

Replace NaN with 0, technically that makes more sense for term frequency number.

```python
tfidftable = tfidftable.fillna(value = 0)
```
