```python
import numpy as np
import pandas as pd
```

```python
all_texts = pd.read_csv('https://raw.githubusercontent.com/niekveldhuis/Digital-Assyriology/test/Scrape-etcsl/Output/alltexts.csv')
```

```python
all_texts = all_texts.loc[all_texts.lang == 'sux', :]
```

```python
all_texts[:20]
```

```python
all_terms = [str(all_texts.cf[i]) + '[' + str(all_texts.gw[i]) + ']' + str(all_texts.pos[i]) for i in all_texts.index]
all_texts['term'] = all_terms
```

```python
#just curious...
unique_terms = list(set(all_terms))
len(unique_terms)
```

```python
### CALCULATING TF ###

#first get counts of all terms in each document
tf = all_texts.pivot_table(aggfunc=pd.Series.nunique, columns='term', index='id_text', values= 'form')
tf = tf.fillna(0)
tf[:10]
```

```python
#find length of all docs
doc_lengths = pd.DataFrame([len(pd.read_table('corpus/' + doc + '.txt').index) for doc in tf.index])
doc_lengths.index = tf.index
```

```python
#normalize all the counts
normalized_tf = tf.divide(doc_lengths[0], axis='index')
normalized_tf[:10]
```

```python
### CALCULATING IDF ###

number_docs = tf.shape[0]
terms = list(tf.columns.values)
term_frequency = [len(set(all_texts.loc[all_texts.term == term, :].id_text)) for term in terms]
idf = [np.log(number_docs / freq) for freq in term_frequency]

print(term_frequency[:5])
idf[:5]
```

```python
idf = pd.DataFrame(idf).transpose()
idf.columns = tf.columns.values
idf
```

```python
### CALCULATING TF-IDF ###

tf_idf = normalized_tf.multiply(idf.loc[0], axis='columns')
tf_idf[:20]
```
