# DTM Function Development

### This function produces a table of text and each text's words (I have not processed/parsed the words yet.) For some reason I was not able to transpose the matrix and return it; it  would not save (not sure if that is the real problem). Did not know how to label rows or columns...

```python
def process(directory_name, num_files, doc_type=".txt"):
    table = pd.DataFrame()
    all_files = os.fsencode(directory_name)
    stop = num_files
    start = 1
    index = 0
    
    for file in os.listdir(all_files):
            filename = os.fsdecode(file)
            if filename.endswith(".txt"): #".txt"
                with open(directory_name + filename, 'r') as text_file:
                    one_doc = text_file.read()
                    one_doc_list = one_doc.split(':')
                    one_doc_list = [word for word in one_doc_list if '[' in word]
                    table["text_" + str(index)] = pd.Series(one_doc_list)
                    num_files += 1
                    index += 1
                if(start >= stop):
                    return table
                start += 1
            else:
                continue
    return table


test_table = process('corpus/', 5)
transposed_test_table = pd.DataFrame.transpose(test_table)
transposed_test_table
```
