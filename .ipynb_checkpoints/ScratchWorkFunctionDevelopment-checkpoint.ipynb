```python
table2 = pd.DataFrame()
directory = os.fsencode('corpus/')
num_files = 1
index = 0
for file in os.listdir(directory):
    while(num_files <= 4):
        filename = os.fsdecode(file)
        if filename.endswith(".txt"): 
            #print(os.path.join(directory, filename))
            with open('corpus/'+ filename, 'r') as text_file:
                one_doc = text_file.read()
                one_doc_list = enmerkar.split(':')
                one_doc_list = [word for word in one_doc_list if '[' in word]
                table2["text_" + str(index)] = pd.Series(one_doc_list)
                index += 1                    
                num_files += 1
        else:
            continue
#table2 = pd.DataFrame.transpose(table)
table2
```

```python
def process(directory_name, num_files, doc_type=".txt"):
    table = pd.DataFrame()
    all_files = os.fsencode(directory_name)
    stop = num_files
    start = 1
    index = 0
    
    for file in os.listdir(all_files):
            filename = os.fsdecode(file)
            if filename.endswith(".txt"): #".txt"
                #print(os.path.join(directory, filename))
                with open(directory_name + filename, 'r') as text_file:
                    one_doc = text_file.read()
                    one_doc_list = one_doc.split(':')
                    one_doc_list = [word for word in one_doc_list if '[' in word]
                    table[str(index)] = pd.Series(one_doc_list)
                    #print(table)
                    num_files += 1
                    index += 1
                if(start >= stop):
                    return table
                start += 1
            else:
                continue
    #print(table)
    return table


test_table = process('corpus/', 4)
#test_table = pd.DataFrame.transpose(test_table)
test_table
```
